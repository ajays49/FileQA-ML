CODE FOR SUMMARIZER AND TRANSLATOR


```python
!pip install transformers
!pip install torch
!pip install PyMuPDF
!pip install googletrans==4.0.0-rc1
import time
from transformers import T5ForConditionalGeneration, T5Tokenizer
from google.colab import files
import torch
import fitz
from googletrans import Translator

def abstractive_summarization(text):
    model_name = "t5-small"
    model = T5ForConditionalGeneration.from_pretrained(model_name)
    tokenizer = T5Tokenizer.from_pretrained(model_name)

    input_ids = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=512, truncation=True)
    summary_ids = model.generate(input_ids, max_length=150, length_penalty=2.0, num_beams=4, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    return summary

def extract_page_texts(pdf_path):
    doc = fitz.open(pdf_path)
    page_texts = []

    for page_num in range(doc.page_count):
        page = doc[page_num]
        page_texts.append(page.get_text())

    return page_texts

# Upload PDF dynamically
uploaded = files.upload()
pdf_path = list(uploaded.keys())[0]

# Extract text from each page of the PDF
page_texts = extract_page_texts(pdf_path)

# Time the summarization process
start_time_summarization = time.time()
# Perform abstractive summarization without translation
page_summaries = []
for page_text in page_texts:
    summary = abstractive_summarization(page_text)
    page_summaries.append(summary)

# Combine individual page summaries into a single summary
full_summary = " ".join(page_summaries)
# End time for summarization process
end_time_summarization = time.time()
time_taken_summarization = end_time_summarization - start_time_summarization

# Save the generated summary to a text file
with open("summary.txt", "w") as f:
    f.write(full_summary)

print(f"Time taken for summarization: {time_taken_summarization:.2f} seconds")

# Ask the user if translation is necessary
translate_required = input("Do you want to translate the document? (yes/no): ").lower()

if translate_required == "yes":
    # Detect the language of the document
    with open(pdf_path, 'rb') as file:
        first_page_text = fitz.open(file).load_page(0).get_text()

    translator = Translator()
    document_language = translator.detect(first_page_text).lang

    # Ask the user for the target language
    target_language = input(f"The document is in {document_language}. Enter the language you want to translate to: ")

   # Time the translation process
start_time_translation = time.time()
# Translate each page
translated_texts = []
for page_text in page_texts:
    if page_text:
        try:
            # Perform translation
            translation_result = translator.translate(page_text, src=document_language, dest=target_language)
            # Check if translation_result is valid
            if translation_result is not None and translation_result.text:
                translated_texts.append(translation_result.text)
            else:
                print(f"Translation failed or returned an unexpected response for the text: {page_text[:30]}...")  # Optional logging
        except Exception as e:
            print(f"Error during translation: {e}")  # Handle the error gracefully
            # You can handle the error further depending on your needs (e.g., skip this text, terminate, etc.)
# Combine individual translated page texts into a single translated document
translated_document = " ".join(translated_texts)

# End time for translation process
end_time_translation = time.time()
time_taken_translation = end_time_translation - start_time_translation

# Save the translated document to a text file
with open("translated_document.txt", "w") as f:
    f.write(translated_document)

print(f"Time taken for translation: {time_taken_translation:.2f} seconds")


# Print the generated summary
print("Generated summary saved as 'summary.txt'.")
```

```python
!pip install transformers
!pip install sentencepiece
!pip install PyMuPDF
!pip install pandas
!pip install python-docx
!pip install googletrans==4.0.0-rc1

from google.colab import files
import io
import mimetypes
import fitz  # PyMuPDF
from transformers import pipeline, AutoModelForQuestionAnswering, AutoTokenizer
from googletrans import Translator

# Load question-answering model
model_name = "deepset/bert-large-uncased-whole-word-masking-squad2"
qa_model = AutoModelForQuestionAnswering.from_pretrained(model_name)
qa_tokenizer = AutoTokenizer.from_pretrained(model_name)
qa_pipeline = pipeline("question-answering", model=qa_model, tokenizer=qa_tokenizer)

# Function to read text content from different file types
def read_file(file_content, file_name):
    file_type, _ = mimetypes.guess_type(file_name)
    # Save the file with a specific filename
    temp_filename = '/content/temp_file.pdf'
    with open(temp_filename, 'wb') as temp_file:
        temp_file.write(file_content)

    if file_type == 'application/pdf':
        # Extract text from PDF using PyMuPDF
        pdf_document = fitz.open(temp_filename)
        text = ''
        for page_num in range(pdf_document.page_count):
            page = pdf_document[page_num]
            text += page.get_text()
        return text
    elif file_type == 'application/vnd.openxmlformats-officedocument.wordprocessingml.document':
        # Extract text from Word document
        doc = Document(io.BytesIO(file_content))
        text = ' '.join([paragraph.text for paragraph in doc.paragraphs])
        return text
    # Add handling for other file types here...

# Function for abstractive summarization
def abstractive_summarization(text):
    model_name = "t5-small"
    model = T5ForConditionalGeneration.from_pretrained(model_name)
    tokenizer = T5Tokenizer.from_pretrained(model_name)

    input_ids = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=512, truncation=True)
    summary_ids = model.generate(input_ids, max_length=150, length_penalty=2.0, num_beams=4, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    return summary

# Function for translation
def translate_text(text, target_language='fr'):
    translator = Translator()
    translated_text = translator.translate(text, dest=target_language).text
    return translated_text

# Interactive loop to ask questions until 'exit' is entered
while True:
    user_question = input("Ask a question ('exit' to stop): ")
    if user_question.lower() == 'exit':
        print("Exiting...")
        break

    # Upload a file
    uploaded_files = files.upload()

    if not uploaded_files:
        print("Please upload a file.")
        continue

    file_name = next(iter(uploaded_files))
    file_content = uploaded_files[file_name]
    context = read_file(file_content, file_name)

    # Get the answer to the question
    answer = qa_pipeline(question=user_question, context=context)['answer']
    print("Answer:")
    print(answer.replace('\n', ' '))  # Replace newlines with spaces for better readability

    # Summarize the context
    page_summaries = []
    page_texts = context.split('\n\n')  # Split context into paragraphs
    for page_text in page_texts:
        summary = abstractive_summarization(page_text)
        page_summaries.append(summary)

    # Combine individual page summaries into a single summary
    full_summary = " ".join(page_summaries)
    print("\nSummary:")
    print(full_summary)

    # Ask if the user wants to translate the content
    request_translation = input("Do you want to translate the content? (yes/no): ").lower()
    if request_translation == 'yes':
        target_language = input("Enter the target language code (e.g., 'fr' for French): ")
        translated_text = translate_text(full_summary, target_language)
        print("\nTranslated Text:")
        print(translated_text)
```
